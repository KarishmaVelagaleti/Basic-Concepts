```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,5, 10)
x
```

```python
y = x**2
```

```python
plt.plot(x,y, 'r')
plt.xlabel('X label')
plt.ylabel('Y Label')
plt.title("Title")
```

```python
fig = plt.figure()
axes = fig.add_axes([1, 1, 1, 1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x, 'r')
axes[1].set_title('Second Plot')
plt.tight_layout()
```

### Figure size

```python
fig = plt.figure(figsize =(7,4))
ax = fig.add_axes([0,0,1,1])
```

```python
fig, axes = plt.subplots(nrows=2, ncols = 1, figsize=(7,4))
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, 'purple', lw = 3, alpha = 0.9,ls = ':', marker = 'o', markersize = 10, markerfacecolor='red',label='X Squared')
ax.plot(x, x**3, 'orange', label = 'X Cubed')

ax.legend(loc=1)
```
